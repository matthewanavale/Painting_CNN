```python
import csv
import requests
from urllib.parse import urlparse
import os
```

```python

def download_image(url, save_path):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(save_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded: {url}")
        else:
            print(f"Failed to download: {url}")
    except Exception as e:
        print(f"Error downloading {url}: {e}")
```

```python
csv_file = 'C:\\Users\\Matthew\\Downloads\\wikiart_scraped.csv'  
output_directory = 'E:\\Data Analytics\\Raw Data\\Surrealism\\'
encoding = 'utf-8'
```

```python
os.makedirs(output_directory, exist_ok=True)

start_row = 42271 # Start downloading from row 161
stop_row = 45870


with open(csv_file, 'r', newline='', encoding=encoding) as file:
    reader = csv.DictReader(file)
    current_row = 1  # Initialize the current row counter
    for row in reader:
        if current_row < start_row:
            current_row += 1  # Increment the current row counter
            continue  # Skip rows until reaching the desired starting row

        image_url = row['Link']  # Use the 'Link' column for image URLs
        if image_url:
            filename = urlparse(image_url).path.split('/')[-1]
            save_path = os.path.join(output_directory, filename)
            download_image(image_url, save_path)

        current_row += 1  # Increment the current row counter

        if current_row > stop_row:
            break  # Exit the loop after reaching the specified stop row
```
