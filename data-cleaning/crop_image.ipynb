```python
from PIL import Image
import os
```

List all the file in the directory

```python
file_path = "D:/TUP SCHOOLWORKS/3rd Year/ACTIVITIES/1ST SEM/DATA ANALYTICS/Painting/my_dataset/Realism_data/realism/"
destination_path = "C:/Users/ASUS/Downloads/crop_realism/"
file_img_list = os.listdir(file_path)
```

Get the total number of images inside the file

```python
print(f"Number of images: {len(file_img_list)}")
```

<strong>Crop</strong> the images

```python
sorted_file_img_list = sorted(file_img_list)
num_img_cropped = 0

# Check if the destination directory exists, and create it if not
if not os.path.exists(destination_path):
    os.makedirs(destination_path)
    
for file_name in sorted_file_img_list:
    print(file_name)
    file_img_path = os.path.join(file_path, file_name)

    try:
    # Crop the image to desired dimension
        image = Image.open(file_img_path)
        width, height = image.size
        left = (width - 640) / 2
        top = (height - 640) / 2
        right = (width + 640) / 2
        bottom = (height + 640) / 2

        print(f"Left: {left}, Right: {right}, Upper: {top}, Lower: {bottom}")
        cropped_image = image.crop((left, top, right, bottom))
        num_img_cropped += 1
        cropped_image.save(os.path.join(destination_path, file_name))
    except Exception as e:
        print(f"Error processing {file_name}: {str(e)}")

# For verification
print(f"\nNumber of standard images: {len(sorted_file_img_list)}")
print(f"Number of cropped images: {num_img_cropped}")
```
